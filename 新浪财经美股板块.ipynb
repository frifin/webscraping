```python
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait
from urllib.parse import quote




import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import time
```

```python
browser = webdriver.Chrome()
wait = WebDriverWait(browser, 10)
```

```python
data_list = []

def index_page(page):

  
    print ('printing page', page)
    
    wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '#data')))
    get_usstock()
    
        
    try:
        submit = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'div#pages>a:last-child')))
        submit.click()
        
    except TimeoutException:
        print('error')
    

def get_usstock():
    time.sleep(5)
    html = browser.page_source
    soup = BeautifulSoup(html, 'html5lib')
    
    for idx, tr in enumerate(soup.find_all('tr')):
        if idx != 0:
            tds = tr.find_all(['td','th'])
            data_list.append({
                '名称': tds[0].text,
                '代码': tds[1].text,
                '最新价': tds[2].text,
                '涨跌额': tds[3].text,
                '涨跌幅': tds[4].text,
                '振幅': tds[5].text,
                '昨收/今开盘': tds[6].text,
                '最高/最低价': tds[7].text,
                '成交量': tds[8].text,
                '市值（亿）': tds[9].text,
                '市盈率': tds[10].text,
                '行业板块': tds[11].text,
                '上市地': tds[12].text
            })
            
    stockdata = pd.DataFrame(data_list)
    #print(stockdata)
    stockdata.to_csv('usstockdata.csv')
    

def main():
    try:
        url = 'http://finance.sina.com.cn/stock/usstock/sector.shtml'
        browser.get(url)
        
    except TimeoutException:
        print('error')
        
        
    MAX_PAGE = 2
    for i in range (1, MAX_PAGE+1):
        index_page(i)
    
    browser.close()
    

if __name__ == '__main__':
    main()
    
```
