```python
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait
from urllib.parse import quote


```

```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import time
```

```python
browser = webdriver.Chrome()
wait = WebDriverWait(browser, 10)
```

```python
data_list = []

def index_page(page):
    print ('printing page', page)
    try:
        url = 'http://quote.eastmoney.com/center/gridlist.html#hk_stocks'
        browser.get(url)
        if page > 1:
            input = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'div#main-table_paginate>input')))
            submit = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'div#main-table_paginate> .paginte_go')))
    
            input.clear()
            input.send_keys(page)
            submit.click()
            
        wait.until(EC.text_to_be_present_in_element((By.CSS_SELECTOR, 'span.paginate_page'), str(page)))
        wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '#table_wrapper-table')))
        
        get_stock()
    except TimeoutException:
        print('error')

def get_stock():
    time.sleep(5)
    html = browser.page_source
    soup = BeautifulSoup(html, 'html5lib')
    
    for idx, tr in enumerate(soup.find_all('tr')):
        if idx != 0:
            tds = tr.find_all('td')
            data_list.append({
                '序号': tds[0].text,
                '代码': tds[1].text,
                '名称': tds[2].text,
                '最新价': tds[4].text,
                '涨跌额度': tds[5].text,
                '涨跌幅': tds[6].text,
                '今开': tds[7].text,
                '最高': tds[8].text,
                '最低': tds[9].text,
                '昨收': tds[10].text,
                '成交量': tds[11].text,
                '成交额': tds[12].text
            })
            stockdata = pd.DataFrame(data_list)
    #print(stockdata)
    stockdata.to_csv('stockdata.csv')
     

    


def main():
    MAX_PAGE = 2
    for i in range (1, MAX_PAGE+1):
        index_page(i)
    
    browser.close()

    

if __name__ == '__main__':
    main()
    

```
